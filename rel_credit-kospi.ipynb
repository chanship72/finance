```python
%load_ext autoreload
%matplotlib inline

import FinanceDataReader as fdr
import matplotlib.pyplot as plt
import time, datetime
import os
import pandas as pd

plt.rcParams["font.family"] = 'nanummyeongjo'
plt.rcParams["figure.figsize"] = (14,4)
plt.rcParams['lines.linewidth'] = 2
plt.rcParams["axes.grid"] = True
```

```python
%autoreload
# KS11 (KOSPI 지수)
df_kospi = fdr.DataReader('KS11', '2000-02-01')
df_kospi['Close'].plot()
```

```python
%autoreload
# KQ11 (KOSDAQ 지수)
df_kosdaq = fdr.DataReader('KQ11', '2000-02-01')
df_kosdaq['Close'].plot()
```

```python
%autoreload

filePath = "equity_credit_history.csv"
data = pd.read_csv(filePath)
datetime_index = pd.to_datetime(data['date'])
df_credit=data.set_index(datetime_index)
df_credit['equity'].plot()
```

```python
df_credit['kosdaq'].plot()
```

```python
# fig = plt.figure(figsize=(14,8))
fig, ax1 = plt.subplots(figsize=(23,8))
ax2 = ax1.twinx()

line1 = ax1.plot(df_credit.index, df_credit['kosdaq'], color='b', linestyle='--', label='KOSDAQ신용거래융자')
line2 = ax2.plot(df_kosdaq.index, df_kosdaq['Close'], color='r', linestyle='--', label='KOSDAQ종가')
lines = line1 + line2
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc=2)
plt.show()
```

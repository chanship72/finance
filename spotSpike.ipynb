```python
%load_ext autoreload
%autoreload
import time, datetime
import os
import pandas as pd
pd.set_option('display.max_colwidth', -1)
```

```python
def get_stockHistory(code, start_date, end_date):
    filePath = "Korea_Stocks_Full/" + str(code).zfill(6) + ".csv"
    data = pd.read_csv(filePath)
    
    datetime_index = pd.to_datetime(data['Date'])
    df=data.set_index(datetime_index)
    df.drop('Date',axis=1,inplace=True)
    
    df = df.loc[start_date:end_date]
    
    return df
```

```python
def addSMA(df,n):
    df['SMA'] = df.Volume.rolling(n, min_periods=n).mean().round()
    result = df[df['Volume'] > df['SMA']]
    
    return result
```

```python
def getStockList(filePath):
    data = pd.read_csv(filePath)
    
    data = data[['name','code']]
    return data
# print(getStockList('total_stocklist.csv'))
```

```python
def getStockNdaySpike(currDate, n):
    date_1 = datetime.datetime.strptime(currDate, "%Y-%m-%d")
    date_2 = date_1 - datetime.timedelta(days=n*2)
    startDate = date_2.strftime("%Y-%m-%d")

    stockList = getStockList('total_stocklist.csv')
    total_result = pd.DataFrame()

    for i in range(len(stockList.index)):
        try:
            sCode = stockList.loc[i]['code']
            sName = stockList.loc[i]['name']
            df = get_stockHistory(sCode, startDate, currDate)
            result = addSMA(df, n)
            if type(result) == pd.core.frame.DataFrame:
                if len(result):
                    result['sCode'] = sCode
                    result['sName'] = sName
                    total_result = pd.concat([total_result, result])

        except Exception as e:
            print(str(e))
            pass
    return total_result
```

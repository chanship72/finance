```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
for i in range(365*20//20):
    html = requests.get('https://finance.naver.com/item/frgn.nhn?code=005930&page={}'.format(i + 1))
    table = pd.read_html(html.text)
    if i == 0:
        data = table[2].dropna().reset_index(drop = True)
    if i > 0:
        data1 = table[2].dropna().reset_index(drop = True)
        data = pd.concat([data, data1])
```

```python
data
```

```python
data = data.reset_index(drop = True)

for i in range(len(data)):
    if float(data['등락률', '등락률'][i].replace('%', '')) < 0 :
        data['전일비','전일비'][i] = -data['전일비', '전일비'][i]

data.to_excel('samsung.xlsx')
```
